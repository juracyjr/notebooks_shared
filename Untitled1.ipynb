```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("/dados/isc/erika_pau_da_lima_ubs/ubs_pontos_geo.csv",sep=";")
```

```python
df
```

```python
df_ok = df.dropna(axis=0, how='any')
```

```python
df_ok
```

```python
m = folium.Map(
    location=[-12.934866, -38.443739],
    zoom_start=13
)

def func_gerar_markers(row):
    if row['tipo'] == "UBS":
        folium.Marker(
            location=[row['lat'], row['long']],
            popup=row['nome'],
            icon=folium.Icon(color='green')
        ).add_to(m)
    elif row['tipo'] == "USF":
        folium.Marker(
            location=[row['lat'], row['long']],
            popup=row['nome'],
            icon=folium.Icon(color='red')
        ).add_to(m)
    else:
        folium.Marker(
            location=[row['lat'], row['long']],
            popup=row['nome'],
            icon=folium.Icon(color='blue')
        ).add_to(m)

df_ok.apply(func_gerar_markers,axis=1)
```
